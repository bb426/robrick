```python
import requests
import pandas as pd
import calendar, time
from datetime import datetime
```

```python
API_HOST = 'https://opensky-network.org/api/flights/arrival?airport='
```

# Arrivals by airport during a given time

```python
def arrival_by_airport_given_time(airport_ICAO, begin_time, end_time):

    # Convert time to Unix-time (Form receiving : yyyy-mm-dd HH:MM:SS)
    def convert_utc_to_epoch(timestamp_string):
        '''Use this function to convert utc to epoch'''
        timestamp = datetime.strptime(timestamp_string, '%Y-%m-%d %H:%M:%S')
        epoch = int(calendar.timegm(timestamp.utctimetuple()))
        return epoch
    
    begin_unix_time = str(convert_utc_to_epoch(begin_time))
    end_unix_time = str(convert_utc_to_epoch(end_time))
    
    url = API_HOST + airport_ICAO + '&begin=' + begin_unix_time + '&end=' + end_unix_time
    res = requests.get(url)    
    
    result = pd.DataFrame(res.json())
    result['firstSeen'] = result['firstSeen'].apply(lambda x: time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(x)))
    result['lastSeen'] = result['lastSeen'].apply(lambda x: time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(x)))
    
    return result
```

```python
result = arrival_by_airport_given_time('LEBL', '2019-01-01 00:00:00', '2019-01-02 00:00:00')
result.head().T
```
